```python
import gradio as gr
import pickle
import numpy as np

with open("Heartattack_prediction.pkl","rb") as f:
    model=pickle.load(f)


def Heartattact_predict_model(*k):
    val=model.predict([k])
    if val==1:
        return "Possibility to get the HeartAttack"
    elif val==0:
        return "Healthy"
    
with gr.Blocks() as demo:
    
    gr.Markdown("# Heart Predicition system")
    
    
    with gr.Row():
        with gr.Column():
            age=gr.Textbox(label="Enter age of the person in years")
            sex=gr.Textbox(label="Enter your Gender (1 = Male, 0 = Female):")
            cp=gr.Textbox(label="Enter your Types of chest pain (0-3), indicating angina levels")
            trestbps=gr.Textbox(label="Enter your Blood pressure (in mm Hg) at rest:")
            chol=gr.Textbox(label="Enter your Cholesterol level in mg/dL:")
            fbs=gr.Textbox(label="Enter your Whether fasting blood sugar is >120 mg/dL (1 = Yes, 0 = No):")
            restecg=gr.Textbox(label="Enter your ECG results (0 = Normal, 1 = ST-T abnormality, 2 = LVH):")
            thalach=gr.Textbox(label="Enter your Maximum heart rate during exercise:")
            exang=gr.Textbox(label="Enter your Chest pain during exercise (1 = Yes, 0 = No):")
            oldpeak=gr.Textbox(label="Enter your Deviation in ECG caused by exercise:")
            slope=gr.Textbox(label="Enter your Measures heart recovery (0-2):")
            ca=gr.Textbox(label="Enter your Count of major vessels (0-4) colored by fluoroscopy:")
            thal=gr.Textbox(label="Enter your Thalassemia 3 = Normal, 6 = Fixed Defect, 7 = Reversible Defect")


            values=[age,sex,cp,trestbps,chol,fbs,restecg,thalach,exang,oldpeak,slope,ca,thal]
    

            greet_btn=gr.Button("Predict")
            output=gr.Textbox(label="OutputBox:")
            greet_btn.click(fn=Heartattact_predict_model,inputs=values,outputs=output,api_name="Prediction")
        with gr.Column():
            gr.Image("F:\Heart_1.png",elem_id="Header-img")
            

demo.launch()
```
